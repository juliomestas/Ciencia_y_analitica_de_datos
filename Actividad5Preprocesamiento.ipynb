5. ¿Tendrán los valores faltantes de `Weight` relación con alguna otra variable? Analízalo considerando las siguientes columnas:
*   Numéricas: `Age, Number of siblings`
*   Categóricas: `Gender, Education`

Para ello deberás dividir cada columna en dos grupos: con/sin faltantes en `Weight` y utilizar métodos gráficos y la prueba t para ver si hay diferencias significativas.

**NOTA**. Puedes reutilizar las funciones `Diagnose_MV_Numerical` y `Diagnose_MV_Categorical` definidas en Hands-On-Prep_Missing_Values.zip. Revisa nuevamente cómo utilizarlas.

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])

    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(10,4))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
# Crear una Serie de booleanos para identificar los registros con valores faltantes en 'Weight'
BM_MV_Weight = response_df['Weight'].isnull()

# Diagnóstico para la variable numérica 'Age'
print("Diagnóstico para 'Age' en relación con los valores faltantes de 'Weight'")
Diagnose_MV_Numerical(response_df, 'Age', BM_MV_Weight)

# Diagnóstico para la variable numérica 'Number of siblings'
print("Diagnóstico para 'Number of siblings' en relación con los valores faltantes de 'Weight'")
Diagnose_MV_Numerical(response_df, 'Number of siblings', BM_MV_Weight)

# Diagnóstico para la variable categórica 'Gender'
print("Diagnóstico para 'Gender' en relación con los valores faltantes de 'Weight'")
Diagnose_MV_Categorical(response_df, 'Gender', BM_MV_Weight)

# Diagnóstico para la variable categórica 'Education'
print("Diagnóstico para 'Education' en relación con los valores faltantes de 'Weight'")
Diagnose_MV_Categorical(response_df, 'Education', BM_MV_Weight)
```

7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
# Gráfico de boxplot para 'Weight'
plt.figure(figsize=(8, 6))
sns.boxplot(y=response_df['Weight'])
plt.title('Boxplot de Weight')
plt.show()
```

```python
# Cálculo del rango intercuartílico (IQR)
Q1 = response_df['Weight'].quantile(0.25)
Q3 = response_df['Weight'].quantile(0.75)
IQR = Q3 - Q1

# Definición de los límites para considerar un valor como atípico
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar y almacenar los valores atípicos en un dataframe independiente
outliers_df = response_df[(response_df['Weight'] < lower_bound) | (response_df['Weight'] > upper_bound)]

# Eliminar los valores atípicos del dataframe original
response_df = response_df[(response_df['Weight'] >= lower_bound) & (response_df['Weight'] <= upper_bound)]

print(f"Número de valores atípicos: {len(outliers_df)}")
```

9. ¿Hay encuestados que sean atípicos según su índice de masa corporal (`BMI`)? El `BMI` puede ser calculado usando la siguiente fórmula:

`BMI = Weight / Height ^ 2`

**NOTA**. El peso tiene que estar en kilogramos y la altura en metros, pero en el conjunto la altura se ha registrado en centímetros, por lo que deberás convertirla a metros.

```python
# Convertir la altura de centímetros a metros
response_df['Height_m'] = response_df['Height'] / 100
```

```python
# Calcular el BMI
response_df['BMI'] = response_df['Weight'] / (response_df['Height_m'] ** 2)
```

```python
# Gráfico de boxplot para BMI
plt.figure(figsize=(8, 6))
sns.boxplot(y=response_df['BMI'])
plt.title('Boxplot de BMI')
plt.show()
```

```python
# Cálculo del rango intercuartílico (IQR) para BMI
Q1_bmi = response_df['BMI'].quantile(0.25)
Q3_bmi = response_df['BMI'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi

# Definición de los límites para considerar un valor como atípico para BMI
lower_bound_bmi = Q1_bmi - 1.5 * IQR_bmi
upper_bound_bmi = Q3_bmi + 1.5 * IQR_bmi

# Identificar y almacenar los valores atípicos de BMI en un dataframe independiente
outliers_bmi_df = response_df[(response_df['BMI'] < lower_bound_bmi) | (response_df['BMI'] > upper_bound_bmi)]

print(f"Número de encuestados con BMI atípico: {len(outliers_bmi_df)}")
```
